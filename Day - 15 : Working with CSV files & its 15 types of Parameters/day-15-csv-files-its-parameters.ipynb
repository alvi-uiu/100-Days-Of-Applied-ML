# 3. Opening csv from an url :

```python
# we have a csv file in this url : https://raw.githubusercontent.com/dr5hn/countries-states-cities-database/refs/heads/master/csv/countries.csv
```

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/dr5hn/countries-states-cities-database/refs/heads/master/csv/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

# 4. Sep Parameter : 

```python
# if we have a tsv (tab-separated-value) dataset then the values may broke : 
pd.read_csv('/kaggle/input/movie-title/movie_titles_metadata.tsv')
```

```python
# so we need to set , sep parameter : 
pd.read_csv('/kaggle/input/movie-title/movie_titles_metadata.tsv',sep='\t')
```

```python
# but the problem here is as this dataset does not have any col name , so it made the first row as a col name :
pd.read_csv('/kaggle/input/movie-title/movie_titles_metadata.tsv', sep='\t',names=['serial no','name','release year','rating','votes','genres'])
```

# 13. Convertors : 

```python
def rename(name):
    if name=='Royal Challengers Bangalore':
        return 'RCB'
    return name
```

```python
# if we want to rename any values from the col : Royal Challengers Bangalore -> RCB
pd.read_csv('/kaggle/input/ipl-data/IPL Matches 2008-2020.csv',converters ={'team1':rename})
```
